```python
import matplotlib.pyplot as plt

def ingresar_valores():
    valores = []
    while True:
        valor = input("Ingresa un valor (o 'fin' para terminar): ")
        if valor.lower() == 'fin':
            break
        try:
            valor = float(valor)
            valores.append(valor)
        except ValueError:
            print("Por favor, ingresa un valor num√©rico v√°lido.")

    return valores

def mostrar_lista_valores(ax, valores):
    ax.text(1.1, 0.5, "Lista de valores ingresados:", transform=ax.transAxes, fontsize=12, verticalalignment='center')
    for i, valor in enumerate(valores, start=1):
        ax.text(1.1, 0.45 - 0.05 * i, f"{i}. {valor}", transform=ax.transAxes, fontsize=12, verticalalignment='center')

def mostrar_graficos(valores):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

    # Gr√°fico de torta
    ax1.pie(valores, autopct='%1.1f%%', startangle=90, colors=['blue', 'orange', 'green', 'red', 'purple'])
    ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    ax1.set_title('Gr√°fico de Torta')

    # Gr√°fico de l√≠neas
    ax2.plot(range(1, len(valores) + 1), valores, marker='o', linestyle='-')
    ax2.set_xlabel('√çndice')
    ax2.set_ylabel('Valor')
    ax2.set_title('Gr√°fico de L√≠neas')

    mostrar_lista_valores(ax2, valores)

    plt.tight_layout()
    plt.show()

def main():
    print("Ingresa valores para los gr√°ficos.")
    valores = ingresar_valores()

    if valores:
        mostrar_graficos(valores)
    else:
        print("No se ingresaron valores para mostrar los gr√°ficos.")

if __name__ == "__main__":
    main()
```
